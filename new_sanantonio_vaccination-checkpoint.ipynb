```python
Data_frame["Date"] = [x[:10] for x in Data_frame["Date"].tolist()]
Data_frame
```

```python
One_Dose = Data_frame.groupby('Date').agg({"One Dose":"sum"})
One_Dose
```

```python
category_data = Data_frame.groupby('Category')
category_data.head()
```

```python
dates = Data_frame["Date"].unique()
Fully_Vaccinated = Data_frame.groupby("Date").agg({"Fully Vaccinated":"sum"})
Fully_Vaccinated
```

```python
plt.plot(dates, One_Dose["One Dose"],color="red", label="One Dose")
plt.plot(dates, Fully_Vaccinated["Fully Vaccinated"],color="blue", label="Fully Vaccinated")
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.title("one dose/Fully Vaccinated over time ")
plt.xlabel("Date")
plt.ylabel("people vaccinted")
plt.show()
```

```python
age_groups_df = Data_frame.loc[Data_frame["Category"] == "5Cat Age Groups"]
age_groups = age_groups_df["Category Description"].unique()
group_sum = [age_groups_df[age_groups_df["Category Description"] == x]["Fully Vaccinated"].sum() for x in age_groups]
group_sum
```

```python
group_sum
x_axis = np.arange(len(group_sum))
```

```python
tick_locations = [value for value in x_axis]
plt.bar(age_groups, group_sum, color='r', alpha=0.5, align="center")
plt.title("Age Group/Fully vaccinted ")
plt.xlabel("Age Group")
plt.ylabel("Number of Fully vaacinted people")
plt.show()
```

```python
one_dose_sum = [age_groups_df[age_groups_df["Category Description"] == x]["One Dose"].sum() for x in age_groups]
one_dose_sum
```

```python
tick_locations = [value for value in x_axis]
plt.bar(age_groups, one_dose_sum, color='b', alpha=0.5, align="center")
plt.title("Age Group/one dose vaccnited ")
plt.xlabel("Age Group")
plt.ylabel("Number of one dose vaacinted people")
```
